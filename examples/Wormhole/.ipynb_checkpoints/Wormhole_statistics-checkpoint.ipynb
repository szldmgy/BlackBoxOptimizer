
Collecting data from result files of the experiment into a single `out.csv`.
The most recent directory is `data2`, the other ones are older and there might by some errors.. 



```python
#collect all data in one single file
import os



directory1 = ["./data1/"]
directory2 = ["./data2/"] #this is the most recent, the others can contain problems
directory0 = ["./data/"]
directory075 = ["./data075/"] #folder for noise = 0.75
directory05 = ["./data05/"] #folder for noise = 0.5
directory03 = ["./data03/"] #folder for noise = 0.3





directory = ["./data1/","./data2/","./data/"]


d = directory03
flist = []
for actdir in d:
    for file in os.listdir(actdir):
        if file.endswith(".csv"):
            flist.append(actdir+ file)

fout=open("out.csv",'w')
# first file:
for line in open(flist[0]):
    fout.write(line)
# now the rest:    
for num in range(0,len(flist)-1):
    try:
        print(flist[num])
        f = open(flist[num])
        fout.write('\n')
        next(f)#.next() # skip the header
        for line in f:
            if line.count(",")==8 :
                fout.write(line)
        f.close() # not really needed
    except Exception as e:
        print(e)
fout.close()

```

Visualization
-
3 plots:
        1. `fig_avg_` split search space into `section_count` slices, take their average and plot that.
        2. `fig_rolled_avg_` applying rolling mean on the last one.
        3. `fig_rolled_` only taking rolling mean, but the distribution might be far from uniform apart from data2

```python
%matplotlib inline
import matplotlib
import pandas as pd
import numpy as np

df = pd.read_csv('out.csv')
#remove failed experiments
df =df[~((df.true_positives == 0) & (df.false_positives == 0) & (df.true_negatives == 0) & (df.false_negatives == 0))]
df[df.true_positives==0 || df.fals_negatives==0].sum() / len(df.index)  


#compute recall for all test
df['recall'] = df['true_positives']/(df['true_positives']+df['false_negatives'])
#sort by average degree
sorted = df.sort_values(by='avg_degree',ascending = True)
# take rolling mean
sorted['rolled'] = sorted['recall'].rolling(100,min_periods=10,center = True).mean()


section_count = 100

#check distribution
min_avg_degree = min(sorted['avg_degree'])
max_avg_degree = max(sorted['avg_degree'])
step = (max_avg_degree - min_avg_degree) / section_count 
ranges = np.arange(min_avg_degree,max_avg_degree, step)
distr = sorted.groupby(pd.cut(sorted['avg_degree'], ranges)).mean()

import time
#plot distribution - taking average per  
title = "Averaged inn interval - Sample size: "+str(len(sorted.index))+", P = "+str(min(sorted['p']))+" -> "+str(max(sorted['p']))
plot1 = distr.plot(kind="line", y = 'recall',x = 'avg_degree' ,figsize=(15,10), title = title)
fig1 = plot1.get_figure()
timestr = time.strftime("%Y_%m_%d-%H_%M_%S")
fig1.savefig("fig_avg_"+timestr+".png")

#plot rolled window on avg - making it more smoth
distr['rolled'] = distr['recall'].rolling(4,min_periods=2,center = True).mean()
title = "Rolling window on avg -  Sample size: "+str(len(sorted.index))+", P = "+str(min(sorted['p']))+" -> "+str(max(sorted['p']))
plot3 = distr.plot(kind="line", x='avg_degree',y='rolled',figsize=(15,10), title = title)
fig3 = plot3.get_figure()

fig2.savefig("fig_rolled_avg_"+timestr+".png")





#plot rolling
title = "Rolling window - Sample size: "+str(len(sorted.index))+", P = "+str(min(sorted['p']))+" -> "+str(max(sorted['p']))
plot2 = sorted.plot(kind="line", x='avg_degree',y='rolled',figsize=(15,10), title = title)
fig2 = plot2.get_figure()

fig2.savefig("fig_rolled_"+timestr+".png")

```
